```python
import pandas as pd
import time
from google.colab import auth
from google.auth import default
import gspread

# 1. authenticate_user - Google Sheets
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
# 1. path/sheetID
FILE_PATH = 'Devices with user details.csv'
SPREADSHEET_ID = '1Gw2Z3vJwt3_5BU678XfTJcd-369yNHqBaMRXnfgcc4s'
CHUNK_SIZE = 5000  # max 5000 low each time，avoid API max

# 2. open_work
sh = gc.open_by_key(SPREADSHEET_ID)
worksheet = sh.get_worksheet(0) # 0 (gid=0)

# 3. clear/cover old_data
worksheet.clear()

print("Start loading data...")

# 4. pandas load Google Sheets
# header=True
first_chunk = True

for chunk in pd.read_csv(FILE_PATH, chunksize=CHUNK_SIZE):
    # from NaN to string or avoid
    chunk = chunk.fillna('')

    # load data
    data = chunk.values.tolist()

    # add first row
    if first_chunk:
        header = chunk.columns.tolist()
        worksheet.append_rows([header] + data)
        first_chunk = False
    else:
        worksheet.append_rows(data)

    print(f"added {len(chunk)} row data...")

    # stop 1 sec to avoid Google API max (Quota)
    time.sleep(1)

print("All Done!")
```
